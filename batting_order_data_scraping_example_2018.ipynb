```python
# Read the data
season_val_2018 = pd.read_csv("season_val/2018.csv")
season_scraped_2018 = pd.read_csv("season_scraped/2018.csv")
```

```python
# Check how many teams
len(set(season_scraped_2018.Team.values))
```

```python
# Cross validate
team_2018_merge = season_scraped_2018.groupby(by="Team")\
                                     .PA.sum().reset_index()\
                                     .merge(season_val_2018[["Tm", 'PA']].rename(columns = {'Tm':'Team', 
                                                                                            'PA':'PA_val'}))
team_2018_merge["diff"] = team_2018_merge.PA - team_2018_merge.PA_val
team_2018_merge
```
